```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'A': [1, 2, 3, 4],
    'B': [1, 3, 3, 1],
    'C': [2, 2, 2, 2],
    'D': [4, 3, 2, 1],
    'E': [3, 3, 2, 1],
    'F': [1, 1, 1, 1]
}

x = np.arange(len(data['A']))  # 生成横坐标位置
width = 0.15  # 柱子的宽度

fig, ax = plt.subplots()
for i, (label, values) in enumerate(data.items()):
    x_pos = x + i * width
    rects = ax.bar(x_pos, values, width, label=label)

# 添加标签，标题和图例
ax.set_ylabel('Y-axis')
ax.set_xlabel('X-axis')
ax.set_title('Comparison of A, B, C, D, E, and F')
ax.set_xticks(x + 2.5 * width)
ax.set_xticklabels(['0', '1', '2', '3'])

# 添加图注
labels = {
    'A': "NP=2, Standard",
    'B': "NP=2, Ours",
    'C': "NP=4, Standard",
    'D': "NP=4, Ours",
    'E': "NP=8, Standard",
    'F': "NP=8, Ours"
}

handles, _ = ax.get_legend_handles_labels()
ax.legend(handles, [labels[label] for label in data.keys()])

# 显示柱状图
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
data = {
    (2, 1): 1529685,
    (2, 2): 1490661,
    (2, 3): 1465264,
    (2, 4): 1519337,
    (4, 1): 2906701,
    (4, 2): 2843233,
    (4, 3): 2813979,
    (4, 4): 2702313,
    (8, 1): 4643249,
    (8, 2): 3973864,
    (8, 3): 3970851,
    (8, 4): 3733107
}

node_nums = [2, 4, 8]
thread_nums = [1, 2, 3, 4]
x = np.arange(len(node_nums))  # 生成横坐标位置
width = 0.15  # 柱子的宽度

fig, ax = plt.subplots()

for i, thread_num in enumerate(thread_nums):
    x_pos = x + i * width
    times = [data[(node_num, thread_num)] for node_num in node_nums]
    ax.bar(x_pos, times, width, label=f'Threads: {thread_num}')

# 添加标签，标题和图例
ax.set_yscale('log')  # 设置纵坐标为对数坐标
ax.set_ylabel('Time of My_MultiThread_Reduce (µs)')
ax.set_xlabel('Node Num')
ax.set_title('Performance Comparison by Node Num and Thread Num')

# 修改横坐标刻度标注
ax.set_xticks(x + 1.5 * width)
ax.set_xticklabels(node_nums)

# 添加图注
ax.legend(title='Thread Num')
# Save the plot as a high-resolution PNG image
plt.savefig('multi_thread_bar_chart.png', dpi=300)
# 显示柱状图
plt.show()
```
